```python
# (c) Beherith mysterme@gmail.com 
arraynum_index = 0
ARRAY_POW2_SIZE = 4
szn = "ARRAY_%02d_POW2_SIZE" % arraynum_index
asv = ('ARRAY_%02d_'%arraynum_index) + 'STATIC_VAR_%03d'

import math
def log2f(v):
    return math.floor(math.log2(v))
```

```python
def declare_staticvars(po2_size):
    res = []
    i = 0
    res.append(f'#if {szn} > {0}')
    res.append(f'   static-var {asv%i}, {asv%(i+1)};')
    res.append(f'#endif')
    i = 2
    for p in range(1, po2_size):
        res.append(f'#if {szn} > {p}')

        res.append(f'   static-var {", ".join([asv%j for j in range(i, 2*i )])};')
        res.append(f'#endif')
        i += 2 ** p

    return res


print ('\n'.join(declare_staticvars(4)))
```

```python

def array_getters(po2_size):
    res = []
    res +=['']
    res +=[f'ARRAY_{arraynum_index:02}_GETTER(index) ' + '{']
    res +=[f'\tif(index == {0}) ARRAY_VALUE = {asv%0};']
    res +=[f'\tif(index == {1}) ARRAY_VALUE = {asv%1};']
    i = 2
    for p in range(1, po2_size):
        res.append(f'#if {szn} > {p}')
        for j in range(i, 2*i ):
            res.append(f'\tif(index == {j}) ARRAY_VALUE = {asv%j};')
        res.append(f'#endif')
        i += 2 ** p

    res += ['}']
    return res
    
print ('\n'.join(array_getters(4)))
```

```python
def recursive_array_getters_2(depth, start, end, res):
    if start +1 == end:
        res += ['\t' * depth + f'ARRAY_VALUE = {asv%start};']

    elif start + 2 == end:
        res += ['\t' * depth + f'if (index == {start}) ARRAY_VALUE = {asv%start};']
        res += ['\t' * depth + f'else ARRAY_VALUE = {asv%(end-1)};']

    else:
        midpoint = (start + end )// 2
        #if midpoint > 2**depth:
        res += ['\t' * depth + f'#if {szn} > {log2f(midpoint)}']

        res += ['\t' * depth + f'if (index < {midpoint}) {{']

        res += ['\t' * depth + f'#endif']
        
        recursive_array_getters_2(depth + 1 , start, midpoint, res)
        
        res += ['\t' * depth + f'#if {szn} > {log2f(midpoint)}']

        res += ['\t' * depth + f'}}else{{']
        recursive_array_getters_2(depth + 1 , midpoint, end, res)
        res += ['\t' * depth +'}']
    
        res += ['\t' * depth + f'#endif']


def array_getters2(po2_size):
    res = []
    res +=['']
    res +=[f'ARRAY_{arraynum_index:02}_GETTER(index) {{' ]

    recursive_array_getters_2(1, 0, 2 ** po2_size, res)
    
    res += ['}']
    return res
    
print ('\n'.join(array_getters2(4)))
```

```python

def array_setters(po2_size):
    res = []
    res +=['']
    res +=[f'ARRAY_{arraynum_index:02}_SETTER(index, value)' + '{']
    res +=[f'\tif(index == {0}) {asv%0} = value;']
    res +=[f'\tif(index == {1}) {asv%1} = value;']
    i = 2
    for p in range(1, po2_size):
        res.append(f'#if {szn} > {p}')
        for j in range(i, 2*i ):
            res.append(f'\tif(index == {j}) {asv%j} = value;')
        res.append(f'#endif')
        i += 2 ** p

    res += ['}']
    return res
    
print ('\n'.join(array_setters(4)))
```

```python
def recursive_array_setters_2(depth, start, end, res):
    if start +1 == end:
        res += ['\t' * depth + f'{asv%start} = value;']

    elif start + 2 == end:
        res += ['\t' * depth + f'if (index == {start}) {asv%start} = value;']
        res += ['\t' * depth + f'else {asv%(end-1)} = value;']

    else:
        midpoint = (start + end )// 2


        res += ['\t' * depth + f'#if {szn} > {log2f(midpoint)}']
        res += ['\t' * depth + f'if (index < {midpoint}) {{'] 
        res += ['\t' * depth + f'#endif']

        recursive_array_setters_2(depth + 1 , start, midpoint, res)


        res += ['\t' * depth + f'#if {szn} > {log2f(midpoint)}']
        res += ['\t' * depth + f'}}else{{']
        recursive_array_setters_2(depth + 1 , midpoint, end, res)
        res += ['\t' * depth +'}']
        res += ['\t' * depth + f'#endif']


def array_setters2(po2_size):
    res = []
    res +=['']
    res +=[f'ARRAY_{arraynum_index:02}_SETTER(index, value) {{' ]

    recursive_array_setters_2(1, 0, 2 ** po2_size, res)
    
    res += ['}']
    return res
    
print ('\n'.join(array_setters2(4)))
```

```python
def make_header(po2_size):
    res = ['//  (c) Beherith mysterme@gmail.com. DO NOT TOUCH: generated by array_header_generator.ipynb']
    res += ['// Notes:']
    res += [f'// #define ARRAY_{arraynum_index:02}_POW2_SIZE']
    res += [f'// use the ARRAY_XX_GET(i) and ARRAY_XX_SET(i,v) function macros']
    res += ['// ARRAY_XX_GET(i) will always place value into ARRAY_VALUE']
    res += ['// WARNING: arrays are not reinitialized on reloadcob!']
    res += ['// average calls per value is about 12']
    res += ['']
    res += [f'#ifndef ARRAY_{arraynum_index:02}_POW2_SIZE']
    res += [f'\t#define ARRAY_{arraynum_index:02}_POW2_SIZE 4']
    res += ['#endif']
    res += ['']
    res += ['#ifndef ARRAY_TEMP_DEFINED']
    res += ['\tstatic-var ARRAY_VALUE;']
    res += ['\t#define ARRAY_TEMP_DEFINED 1;']
    res += ['#endif']
    res += ['']
    res +=[f'#define ARRAY_{arraynum_index:02}_SET(i,v) call-script ARRAY_{arraynum_index:02}_SETTER(i,v);']
    res += ['']
    res +=[f'#define ARRAY_{arraynum_index:02}_GET(i) call-script ARRAY_{arraynum_index:02}_GETTER(i);']
    res += ['']
    res += declare_staticvars(po2_size)
    res += array_getters2(po2_size)
    res += array_setters2(po2_size)
    return res

for i in range(8):
    with open (f"array_{i:02}.h", 'w') as f:
        arraynum_index = i
        szn = "ARRAY_%02d_POW2_SIZE" % i
        asv = ('ARRAY_%02d_'%i) + 'STATIC_VAR_%03d'
        f.write('\n'.join(make_header(8)))
    
```
